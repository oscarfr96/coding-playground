## CLASES Y OBJETOS

```
//Clase
class Sale 
{
    public string ProductName { get; set; }
    public decimal Price { get; set; }
    public int Quantity { get; set; }
    public decimal Total => Price * Quantity;
    private decimal _precioPrivado;

    //Constructor
    public Sale(decimal Price)
    {
        this.Price = Price;
        _precioPrivado = Price*2;
    }

    //Método
    //virtual: permite que un método sea sobrescrito en una clase derivada. Si un método es declarado como virtual, las clases que heredan de la clase base pueden proporcionar su propia implementación de ese método utilizando la palabra clave override.
    public virtual string GetSaleInfo()
    {
        return $"Product: {ProductName}, Price: {Price}, Quantity: {Quantity}, Total: {Total}";
    }

    //protected: solo puede ser accedido por la clase que lo define y por las clases que heredan de ella. No puede ser accedido desde fuera de estas clases.
    protected void ProtectedMethod()
    {
        Console.WriteLine("This is a protected method.");
    }
}

//Objetos (instanciación de una clase)
Sale sale = new Sale(20m);
// var, no como el de javaascript, es un tipo de dato implícito, el compilador infiere el tipo de dato a partir del valor asignado. En este caso, sale es de tipo Sale.
var sale2 = new Sale(20m);
Sale sale3 = new(20m);

var message = sale.GetSaleInfo();
Console.WriteLine(message);
```

# Herencia


```
//Herencia
class SaleWithTax : Sale
{
    public decimal Tax { get; set; }

    //Constructor con herencia, se llama al constructor de la clase base (Sale) para inicializar el precio.
    //El constructor de la clase derivada (SaleWithTax) recibe un parámetro Price y lo pasa al constructor de la clase base (Sale) utilizando la sintaxis : base(Price). Esto asegura que el precio se inicialice correctamente en la clase base.
    public SaleWithTax(decimal Price) : base(Price)
    {
    }

    public string GetInfoWithTax()
    {
        decimal totalWithTax = Total + Tax;
        return $"Product: {ProductName}, Price: {Price}, Quantity: {Quantity}, Total: {Total}, Tax: {Tax}, Total with Tax: {totalWithTax}";
    }

    //Override: es la capacidad de una clase derivada para proporcionar una implementación específica de un método que ya está definido en su clase base. En este caso, se ha sobrescrito el método GetSaleInfo para incluir información adicional sobre el impuesto (Tax) en la salida.
    public override string GetSaleInfo()
    {
        return $"Total: {Total}, Tax: {Tax}";
    }

    //Sobrecarga: es la capacidad de definir múltiples métodos con el mismo nombre pero con diferentes parámetros. En este caso, se ha sobrecargado el método GetSaleInfo para aceptar un parámetro string message. Esto permite proporcionar información adicional al método, como un mensaje personalizado.
    public string GetSaleInfo(string message)
    {
        return $"Message: {message}, Total: {Total}, Tax: {Tax}";
    }
}
```

# Interfaces


```
//Interfaces: es un contrato que define un conjunto de métodos y propiedades que una clase debe implementar. En este caso, se ha definido la interfaz IDiscountable, que requiere que cualquier clase que la implemente proporcione una implementación para el método ApplyDiscount.
//Permiten crear código más escalable, reutilizable y mantenible, ya que las clases pueden implementar múltiples interfaces y proporcionar diferentes funcionalidades sin estar limitadas por una jerarquía de herencia rígida.
interface ISale
{
    decimal Total { get; }
}
interface ISaveable
{
    void Save();
}

public class Sale : ISale, ISaveable
{
    public decimal Total { get; set; }

    public Sale(decimal total)
    {
        Total = total;
    }

    public void Save()
    {
        Console.WriteLine("Sale saved.");
    }
}
```

# Generics


```
//Generics: permiten definir clases, interfaces y métodos con un tipo de dato genérico. Esto permite crear código más flexible y reutilizable, ya que el mismo código puede trabajar con diferentes tipos de datos sin necesidad de duplicar el código para cada tipo específico.
//Muy comunes en colecciones (arrays, listas, etc)
public class MyList<T>
{
    private List<T> _list;
    private int _limit;

    public MyList(int limit)
    {
        _list = new List<T>();
        _limit = limit;
    }

    public void Add(T item)
    {
        if (_list.Count < _limit)
        {
            _list.Add(item);
        }
        else
        {
            Console.WriteLine("List limit reached.");
        }
    }
}

var numbers = new MyList<int>(10);
var strings = new MyList<string>(10);
var sales = new MyList<Sale>(10);
```

# Serializacion y deserializacion


```
//Serializacion y deserializacion: es el proceso de convertir un objeto en un formato que pueda ser almacenado o transmitido, y luego reconstruir ese objeto a partir de ese formato. En C#, se pueden utilizar bibliotecas como System.Text.Json o Newtonsoft.Json para realizar la serialización y deserialización de objetos a JSON, XML u otros formatos.
using System.Text.Json;
public class People
{
    public string Name { get; set; }
    public int Age { get; set; }
    //Azucar sintactico: es una característica del lenguaje de programación que proporciona una sintaxis más conveniente o legible para realizar ciertas tareas, sin agregar nueva funcionalidad al lenguaje. En este caso, el método Get() es un ejemplo de azúcar sintáctico, ya que proporciona una forma más fácil y legible de obtener la información de la persona en lugar de acceder directamente a las propiedades Name y Age.
    public static string Get() => $"Name: {Name}, Age: {Age}";
}

var oscar = new People { Name = "Oscar", Age = 30 };

//Serialización: el proceso de convertir un objeto en un formato que pueda ser almacenado o transmitido. En este caso, se está convirtiendo el objeto oscar en una cadena JSON utilizando el método JsonSerializer.Serialize. El resultado es una representación en formato JSON del objeto oscar, que puede ser almacenada o transmitida fácilmente.
//Serializer es un *metodo estatico*: no es necesario crear una instancia de la clase para usarlo, se puede llamar directamente desde la clase. En este caso, se llama a JsonSerializer.Serialize para convertir el objeto oscar en una cadena JSON.
string json = JsonSerializer.Serialize(oscar);
Console.WriteLine(json);
string myJson = @"
{
    ""Name"": ""Oscar"",
    ""Age"": 30
}";
// Deserialización: el proceso de reconstruir un objeto a partir de su representación en formato JSON. En este caso, se está utilizando el método JsonSerializer.Deserialize para convertir la cadena JSON myJson de nuevo en un objeto de tipo People. El resultado es un nuevo objeto deserializedOscar que contiene los mismos datos que el objeto original oscar.
// Con ? se indica que el tipo de dato puede ser nulo, es decir, que puede contener un valor o no. En este caso, se está indicando que el resultado de la deserialización puede ser un objeto de tipo People o null, lo que permite manejar situaciones en las que la cadena JSON no contenga los datos necesarios para crear un objeto válido.
People? deserializedOscar = JsonSerializer.Deserialize<People>(myJson);
Console.WriteLine(deserializedOscar.Get());
Console.WriteLine(deserializedOscar?.Name);

```

# Programación funcional


```
//Función pura: siempre lo mismo
int Sub(int a, int b)
{
    return a - b;
}

//Función impura: no siempre lo mismo, depende de factores externos
DateTime GetTomorrow()
{
    return DateTime.Now.AddDays(1);
}
//Se convierte en pura si se le pasa la fecha como parámetro, así no depende de factores externos y siempre devuelve lo mismo para la misma entrada.
DateTime GetTomorrow(DateTime date)
{
    return date.AddDays(1);
}

// ---------------------------------------------------

void Show(string message)
{
    Console.WriteLine(message);
}

//Funcion de primera clase: es una función que se trata como un valor, lo que significa que puede ser asignada a una variable, pasada como argumento a otra función o devuelta desde una función. En este caso, la función Show es una función de primera clase porque se puede asignar a la variable show y luego llamar a esa variable como si fuera la función original.
var show = Show;
show("Hello, World!");

//Funcion de orden superior: es una función que toma otra función como argumento o devuelve una función como resultado. En este caso, la función Some es una función de orden superior porque toma una función (Action<string>) como argumento y la ejecuta dentro de su cuerpo.
// Si no retorna nada: Action
void Some(Action<string> action, string message)
{
    Console.WriteLine("Before action");
    action(message);
    Console.WriteLine("After action");
}
//Función de orden superior con función que devuelve un valor: en este caso, la función Some es una función de orden superior que toma una función (Func<string, string>) como argumento y la ejecuta dentro de su cuerpo, además de manejar el resultado devuelto por esa función.
// Si la función retorna algo: Func
void Some(Func<string, string> func, string message)
{
    Console.WriteLine("Before function");
    var result = func(message);
    Console.WriteLine($"Result: {result}");
    Console.WriteLine("After function");
}

```

# LinQ


```
//LinQ (Language Integrated Query): es una característica de C# que permite realizar consultas sobre colecciones de datos de manera más sencilla y legible. LinQ proporciona una sintaxis similar a SQL para consultar datos en colecciones como arrays, listas, etc. En este caso, se está utilizando LinQ para filtrar una lista de números y obtener solo aquellos que son mayores que 5.
var names = new List<string> { "Oscar", "Ligia", "Borja", "Silvia" };
var namesResult = from n in names
                    where n.StartsWith("S")
                    orderby n descending
                    select n;
//La consulta se ejecuta cuando se itera sobre el resultado, en este caso, al usar un foreach para imprimir los nombres filtrados y ordenados. LinQ utiliza la evaluación diferida, lo que significa que la consulta no se ejecuta hasta que se necesite el resultado, lo que permite optimizar el rendimiento al evitar operaciones innecesarias.
foreach (var name in namesResult)
{
    Console.WriteLine(name);
}

//.ToList(): es un método de extensión que se utiliza para convertir el resultado de una consulta LinQ en una lista. En este caso, se está utilizando .ToList() para convertir el resultado de la consulta LinQ (namesResult) en una lista de strings, lo que permite realizar operaciones adicionales sobre esa lista, como agregar o eliminar elementos, o simplemente para tener una colección más concreta y manipulable.
//En colecciones muy grandes, es recomendabqle no usar .ToList() a menos que sea necesario, ya que puede consumir más memoria al crear una nueva lista con todos los elementos del resultado de la consulta. En su lugar, se puede iterar directamente sobre el resultado de la consulta LinQ sin convertirlo a una lista, lo que permite procesar los elementos de manera más eficiente.
var namesList = namesResult.ToList();

// También se puede escribir la consulta LinQ utilizando métodos de extensión en lugar de la sintaxis de consulta, lo que puede ser más conciso y fácil de leer para algunos desarrolladores. En este caso, se está utilizando los métodos Where y OrderByDescending para filtrar y ordenar la lista de nombres, respectivamente, y luego se convierte el resultado a una lista con ToList().
var namesResult2 = names.Where(n => n.StartsWith("S")).OrderByDescending(n => n).ToList();
```
